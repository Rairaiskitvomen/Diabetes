Importing Dependecies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm
```

Data collection and preprocessing

```python
df = pd.read_csv("diabetes.csv")
df.head(6)
```

```python
#checking the shape of the data(Rows and Columns)
df.shape
```

```python
#Getting the statistical summary
df.describe()
```

```python
df.Outcome.value_counts()

# 0 --> Non_Diabetic
# 1 --> Diabetic
```

```python
df.groupby("Outcome").mean()
```

```python
#Seperating the Data from the labels

X = df.drop(["Outcome"] , axis = 1)
Y = df.Outcome
```

The reason why i am standardizing the data is for better prediction by our model, i dont want the values of our features to be different i.e tens , hundreds , thousands so why not make them have same value

```python
scaler = StandardScaler()
```

```python
standardised_data = scaler.fit_transform(X)

# standardised_data is out new X
```

```python
X = standardised_data # to make it easier to call by standardised_data when needed. It is optional
```

Train and splitting the data

```python
X_train, X_test , Y_train, Y_test = train_test_split(X,Y, stratify = Y, test_size = 0.25 , random_state = 1)
```

```python
print(X.shape , X_train.shape , X_test.shape)
```

Training the model

```python
classifier = svm.SVC(kernel = "linear")
```

```python
classifier.fit(X_train , Y_train)
```

```python
#accuracy score on training data
prediction = classifier.predict(X_train)
training_accuracy = accuracy_score(prediction , Y_train)
```

```python
print(training_accuracy)
```

```python
#accuracy score on test data
prediction = classifier.predict(X_test)
test_accuracy = accuracy_score(prediction , Y_test)
```

```python
print(test_accuracy)
```

Making a predictive system

```python
input_array = (7,161,86,0,0,30.4,0.165,47)

#changing the input data as numpy array for faster computation
numpy_input = np.asarray(input_array)

#reshaping for predicting one instance
reshaped_array = numpy_input.reshape(1 , -1)

#standardinzing the input data sice we used standard scaler previously
std_data = scaler.transform(reshaped_array)

Prediction = classifier.predict(std_data)
print(Prediction)

if (Prediction[0] == 0):
  print("the person is not diabetic")
else:
  print("The person is diabetic")
```
